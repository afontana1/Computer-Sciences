```python
def difference_k(arr, target):
    """
    given set of integers, find the pairs yield difference k
    """
    tups = []
    out = {}
    for i in arr:
        y = i - target
        if y not in out:
            out.update({i: True})
        else:
            tups.append((y, i))
    return tups
```

```python
def triplets(arr, target):
    """
    O(N^2)+O(N*Log(N))
    space: O(N)
    """
    gimme = []
    arr_sort = sorted(arr)
    for k in range(len(arr_sort) - 2):
        i = k + 1
        j = len(arr_sort) - 1
        while i < j:
            temp = arr_sort[k] + arr_sort[i] + arr_sort[j]
            if temp == target:
                gimme.append([arr_sort[k], arr_sort[i], arr_sort[j]])
                i += 1
                j -= 1
            if temp > target:
                j -= 1
            if temp < target:
                i += 1
    return gimme


def inefficient_triplets(arr, target):
    """
    O(N^3)
    """
    gimme = []
    for i in range(len(arr) - 2):
        for j in range(i + 1, len(arr)):
            for k in range(j + 1, len(arr)):
                temp = arr[i] + arr[j] + arr[k]
                if temp == target:
                    gimme.append([arr[i], arr[j], arr[k]])

    sorted_ones = []
    for triplet in gimme:
        sorted_ones.append(sorted(triplet))

    unique = []
    for stuff in sorted_ones:
        if stuff not in unique:
            unique.append(stuff)
    return unique
```

```python
from functools import reduce


def max_profit(arr):
    """
    given a price path, determine optimal trading strategy
   
    Input: []
    output: type = int, max price
   
    constraints: Can only buy one share, but can sell as many shares as you want
    """
    stuff = []
    # check if descending
    if all(arr[i] > arr[i + 1] for i in range(len(arr) - 1)):
        return 0
    # if strictly increasing, buy shares on all days
    if all(arr[i] < arr[i + 1] for i in range(len(arr) - 1)):
        return len(arr[:-1]) * arr[-1] - reduce(lambda x, y: x + y, arr[:-1])

    def FindMaxima(numbers):
        maxima = []
        length = len(numbers)
        if length >= 2:
            # check beginning of array
            if numbers[0] > numbers[1]:
                maxima.append(numbers[0])
            # check mid
            if length > 3:
                for i in range(1, length - 1):
                    if numbers[i] > numbers[i - 1] and numbers[i] > numbers[i + 1]:
                        maxima.append(numbers[i])
            # check the end of array
            if numbers[length - 1] > numbers[length - 2]:
                maxima.append(numbers[length - 1])
        return maxima

    maximums = FindMaxima(arr)

    def recursive_sum(arr):
        if len(arr) == 1:
            return arr[0]
        return arr[0] + recursive_sum(arr[1:])

    chunks = []
    for i in maximums:
        idx = arr.index(i)
        newlist = arr[: idx + 1]
        chunks.append(newlist)
        del arr[: idx + 1]

    tot = 0
    for chonk in chunks:
        tot += len(chonk[:-1]) * chonk[-1] - reduce(lambda x, y: x + y, chonk[:-1])

    return tot
```

```python
def shift_one_left(arr):
    """
    utility to shift elements once in array
    returns shifted array
    """
    leftmost = arr[0]
    right_elements = arr[1 : len(arr)]
    return right_elements + [leftmost]


def left_rotate(arr, d):
    """
    Takes array of length N
    rotates the elements leftwards d times
    """
    count = 0
    while count < d:
        temp = shift_one_left(arr)
        arr = temp
        count += 1

    return arr


def sparse_arrays_naive(strings, query):
    """Find the number of occurences a string occurs in array 'strings'
   
    input: list of query strings, list of strings
    output: list of counts each query occurs in array of strings
    example:
        strings: [aba,baba,aba,xzxb]
        query: [aba, xzxb, ab]
        output = [2,1,0]
    """
    out = []
    for q in query:
        temp = 0
        for s in strings:
            if q == s:
                temp += 1
        out.append(temp)

    return out


def sparse_arrays_efficient(strings, query):
    """
    a more efficient version
    """
    out = {}
    for s in strings:
        if s not in out:
            out[s] = 1
        else:
            out[s] += 1

    counts = []
    for q in query:
        if q not in out:
            counts.append(0)
        else:
            counts.append(out[q])

    return counts
```

```python
def specialArraySum(arr,m=1):
    '''A special array is an array that contain other special arrays
    the array will contain integers and potentially additional arrays
    
    The objective is to sumprod the elements in the array, ie sum them up and
    multiply by the length of the array
    '''
    tot = 0
    for element in arr:
        if isinstance(element,list):
            tot+=specialArraySum(element,m+1)
        if isinstance(element,int):
            tot+=element
    return tot*m
```

```python
# code to create a histogram
def sort(arr):
    """Naive sort array"""
    for i in range(len(arr)):
        for j in range(len(arr) - 1):
            if arr[j] > arr[i]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr


def chunk(arr, chunks):
    """Chunk an array and number of chunks"""
    """Similar to histogram but different"""
    cutoffs = int(len(arr) / chunks)
    arr_sort = sort(arr)
    groups = [arr_sort[i : i + cutoffs] for i in range(0, len(arr), cutoffs)]
    return groups


def bin_ranges(arr, bins):
    """Get the cutoffs for an array"""
    """Specify the number of bins you want"""
    array = []
    split = round(len(arr) / bins, 1)
    first = 0
    for i in range(0, bins):
        second = first + split
        array.append(second)
        first = second
    return array


def get_less_than(arr, num):
    """Get values less than a given number"""
    """Helper function used in make_groups"""
    """Assumes right open endpoints"""
    return [i for i in arr if i <= num]


def make_groups(arr, bin_ranges):
    """take in an array and list of cutoffs"""
    """output an array containing groups"""
    out = []
    newarr = arr[:]  # make a copy
    sort_arr = sort(newarr)
    for i in bin_ranges:
        bin_n = get_less_than(sort_arr, i)
        out.append(bin_n)
        for i in bin_n:
            sort_arr.remove(i)
    return out


def get_histogram(arr, bins):
    """return the histogram for an array"""
    if bins < 0:
        raise ValueError("Enter positive integer for bin")
    ranges = bin_ranges(arr, bins)
    groups = make_groups(arr, ranges)
    return [len(bin_) for bin_ in groups]


import random

arr = [random.randint(0, 200) for i in range(200)]
sort(arr)[:20]
# when n---> infinity the histogram distribution should be normal
get_histogram(arr, 4)
```

```python
def oddnumbers(l,r):
    '''
    Given array, print all of the odd numbers
    '''
    nums = []
    for num in range(l,r+1):
        if num%2 != 0:
            nums.append(num)
    return nums

strings = ["we promptly judged antique ivory buckles for the next prize",
            "we promptly judged antique ivory buckles for the prizes",
            "the quick brown fox jumps over the lazy dog",
            "the quick brown fox jump over the lazy dog"]

def pangrams(strings):
    '''
    check if each string is a pangram
    if its not,return 0 , if it is, return 1, join them into one string
    '''
    final = []
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    for st in strings:
        out = {}
        for letter in st.replace(" ","").strip():
            if letter not in out:
                out[letter] = 1
            else:
                out[letter] +=1
                
        if len(out.keys()) == len(alphabet):
            final.append('1')
        else:
            final.append('0')
    return ''.join([i for i in final])

pangrams(strings)
```
